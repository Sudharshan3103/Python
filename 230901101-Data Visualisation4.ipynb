```python
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px

# Define data for 3D scatter plot
X = np.array([1, 2, 6, 8, 12, 14, 20])
Y = np.array([3, 8, 1, 10, 12, 16, 18])
Z = np.array([10, 80, 100, 120, 150, 180, 250])

# Create 3D scatter plot using matplotlib
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(projection='3d')
plt_3d = ax.scatter3D(X, Y, Z, c=Z)
fig.colorbar(plt_3d, ax=ax)
plt.show()

# Define data for 2D scatter plot
X = np.array([10, 80, 100, 120, 150, 180, 250])
Y = np.array([3, 8, 1, 10, 12, 16, 18])

# Create 2D scatter plot using plotly
fig = px.scatter(x=X, y=Y)
fig.show()
```
